```python
import pandas as pd
import numpy as np
```

```python
inc = pd.read_csv('income_per_person_gdppercapita_ppp_inflation_adjusted.csv')
inc.head()
```

```python
le = pd.read_csv('life_expectancy_years.csv')
le.head()
```

```python
pop = pd.read_csv('population_total.csv')
pop.head()
```

```python
incm = inc.melt(id_vars='country', var_name='year', value_name='income')
lem = le.melt(id_vars='country', var_name='year', value_name='life_expect')
popm = pop.melt(id_vars='country', var_name='year', value_name='population')
```

```python
df = pd.merge(incm, lem, how='inner', on=['country','year'])
df = pd.merge(df, popm, how='inner', on=['country', 'year'])
df.head()
```

```python
df = df[df['year'].astype(int) >= 1930]
df = df[df['year'].astype(int) <= 1960]
```

```python
countries = ['Afghanistan', 'Albania', 'Belgium', 'Croatia', 'Spain', 'Mexico', 'Poland', 'Brazil', 'Russia', 'Ukraine', 'Greece', 'England','Zimbabwe']
```

```python
qw = df[df['country'].isin(countries)]
```

```python
aw = qw.sort_values(by=['country', 'year'])
```

```python
aw.to_csv('gapminder_extract.csv', index=False)
```
